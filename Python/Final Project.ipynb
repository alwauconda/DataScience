```python
import pandas as pd
import altair as alt

cfbdata= pd.read_csv("cfb_data.csv")

# Dollar signs are giving us trouble, lets convert to ints
cfbdata["Football Expenses"] = cfbdata["Football Expenses"].replace("[$,]", "", regex=True).astype(float)
cfbdata["Football Revenue"] = cfbdata["Football Revenue"].replace("[$,]", "", regex=True).astype(float)
cfbdata["Football Profit"] = cfbdata["Football Profit"].replace("[$,]", "", regex=True).astype(float)

cfbdata["Basketball Expenses"] = cfbdata["Basketball Expenses"].replace("[$,]", "", regex=True).astype(float)
cfbdata["Basketball Revenue"] = cfbdata["Basketball Revenue"].replace("[$,]", "", regex=True).astype(float)
cfbdata["Basketball Profit"] = cfbdata["Basketball Profit"].replace("[$,]", "", regex=True).astype(float)

#So are percentages

cfbdata["Win Rate"] = cfbdata["Win Rate"].replace("[%]", "", regex=True).astype(float)


cfbdata.head()
```

```python
selection = alt.selection(type='multi', fields=['Conference'])

football = alt.Chart(cfbdata).mark_circle().encode(
    x = alt.X("Football Expenses", title="Football Expenses (Dollars)"),
    y = alt.Y("Football Revenue", title="Football Revenue (Dollars)"),
    color=alt.Color('Conference', scale=alt.Scale(scheme='spectral')),
    size="Football Profit",
    tooltip=["School", "Football Revenue", "Football Profit"],
    opacity=alt.condition(selection,alt.value(1),alt.value(.2))
).add_selection(selection)

basketball = alt.Chart(cfbdata).mark_circle().encode(
    x = alt.X("Basketball Expenses", title="Basketball Expenses (Dollars)"),
    y = alt.Y("Basketball Revenue", title="Basketball Revenue (Dollars)"),
    color=alt.Color('Conference', scale=alt.Scale(scheme='spectral')),
    size="Basketball Profit",
    tooltip=["School", "Basketball Revenue", "Basketball Profit"],
    opacity=alt.condition(selection,alt.value(1),alt.value(.2))
).add_selection(selection)

football | basketball

#(football | basketball).save('sports_revenue.html') 

```
